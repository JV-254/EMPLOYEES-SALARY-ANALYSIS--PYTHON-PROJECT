```python
import pandas as pd
# Creating a dataset
Employee={
    'Employee_ID':['E001','E002','E003','E004','E005','E006','E007','E008','E009','E010','E011','E012'],
    'Name':['John Kamau','Mary Wanjiku','Peter Otieno','Grace Njeri','David Kiptoo','Sarah Achieng',
            'James Mwangi','Linda Chebet','Brian Maina','Faith Atieno','Kevin Mutua','Ruth Wambui'],
    'Department':['IT','HR','IT','HR','Finance','Finance','Marketing','IT','Finance','Marketing','HR','Marketing'],
    'Age':[25,30,35,28,32,40,27,29,38,26,31,36],
    'Experience':[2,5,8,4,6,12,3,5,10,2,7,9],
    'Salary':[55000,65000,85000,70000,60000,120000,50000,75000,110000,58000,80000,95000]
    }
df=pd.DataFrame(Employee)
df

# EXPLORING THE DATASET
# Reading the first 5 rows
ED=df.head()
print(ED)

# Reading the last 5 rows
print('-------------------------------------------------------------------------------------------------------')
ED1=df.tail()
print(ED1)

# Viewing the dataset Information
print('-------------------------------------------------------------------------------------------------------')
ED2=df.info()
print(ED2)

# Viewing dataset description
print('-------------------------------------------------------------------------------------------------------')
ED3=df.describe()
print(ED3)

# Identifying the dataset shape
print('-------------------------------------------------------------------------------------------------------')
ED4=df.shape
print(ED4)

# Calculating the number of Employess
print('-------------------------------------------------------------------------------------------------------')
ED5=df.shape[0]
print(f'Number of Employees: {ED5}')

# Employees average Salary
print('-------------------------------------------------------------------------------------------------------')
ED6=df['Salary'].mean().round(2)
print(f'Employees Average Salary: {ED6}')

# Highest Salary
print('-------------------------------------------------------------------------------------------------------')
ED7=df['Salary'].max()
print(f'Highest Salary: {ED7}')

# Lowest Salary
print('-------------------------------------------------------------------------------------------------------')
ED8=df['Salary'].min()
print(f'Lowest Salary: {ED8}')

# Salary Statistics
print('-------------------------------------------------------------------------------------------------------')
ED9= df["Salary"].describe()
print(f'Salary Statistics: \n  {ED9}')

# DEPARTMENT ANALYSIS
# Average Salary Per Department
print('-------------------------------------------------------------------------------------------------------')
ED10=df.groupby('Department')['Salary'].mean().round(2)
print(f'Average Salary Per Department: \n {ED10}')

# Total Salary Per Department
print('-------------------------------------------------------------------------------------------------------')
ED11=df.groupby('Department')['Salary'].sum()
print(f'Total Salary Per Department: \n {ED11}')

# Number Of Employees Per department
print('-------------------------------------------------------------------------------------------------------')
ED12=df.groupby('Department')['Name'].count()
print(f'Number Of Employees per Department: \n {ED12}')

# EMPLOYEE ANALYSIS
# Highest paid Employee
print('-------------------------------------------------------------------------------------------------------')
ED13=df['Salary'].max()
ED14=df[df['Salary']==ED13]
print(f'Highest paid Employee: \n {ED14}')

# Lowest Paid Employee
print('-------------------------------------------------------------------------------------------------------')
ED15=df['Salary'].min()
ED16=df[df['Salary']==ED15]
print(f'Lowest Paid Employee: \n {ED16}')

# Top 5 Highest paid Employees
print('-------------------------------------------------------------------------------------------------------')
ED17=df.nlargest(5,'Salary')
print(f'Top 5 Highest paid Employees: \n {ED17}')

# FILTERING
# Employees Earning more than 80,000
print('-------------------------------------------------------------------------------------------------------')
ED18=df[df['Salary']>80000]
print(f'Employees Earning more than 80000: \n {ED18}')

# IT Department Employees
print('-------------------------------------------------------------------------------------------------------')
ED18=df[df['Department']=='IT']
print(f'IT Department Employees: \n {ED18}')

# Employees with more than 5 Years of Experience
print('-------------------------------------------------------------------------------------------------------')
ED19=df[df['Experience']>5]
print(f'Employees with Experience More than 5 Years: \n {ED19}')

# SALARY CATEGORIES
# Create function
def category(salary):
    if salary>100000:
        return 'Executive'
    elif salary>80000:
        return 'Senior'
    elif salary>60000:
        return 'Mid-Level'
    else:
        return 'Junior'

# Applying Function to the Dataset
df['Salary Category']=df['Salary'].apply(category)
df

# Department with the Highest average Salary
print('-------------------------------------------------------------------------------------------------------')
ED20=df.groupby('Department')['Salary'].mean().idxmax()
print(f'Deparment With the Highest Average Salary: \n {ED20}')

# Percentage of employees in each department
print('-------------------------------------------------------------------------------------------------------')
ED21=df.groupby('Department')['Name'].count()
ED22=df['Name'].count()
ED23=ED21/ED22*100
print(f'Percentage Of Employees in Each Department: \n {ED23}')

# Employee whose salary is above the company average
ED24=df['Salary'].mean()
ED25=df[df['Salary']>ED24]
print(f'Employee Whose Salary is above Company Salary: \n {ED25}')

# AGE GROUP ANALYSIS
# Create a Function
def Group(age):
    if age >=20 and age<=29:
        return 'Young'
    elif age>=30 and age<=39:
        return 'Adult'
    elif age>=40:
        return 'Senior'
    
# Applying The function to Dataset
df['Age Group']=df['Age'].apply(Group)
df

# Ranking Employees From the Highest to the Lowest
df=df.sort_values(by='Salary', ascending=False)
df
```
